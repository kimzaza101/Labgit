```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df_iris = pd.read_csv('IrisSpecies.csv')

df_iris.plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm')
df_iris.plot(kind='scatter', x='PetalLengthCm', y='PetalWidthCm')
df_iris.plot(kind='scatter', x='SepalLengthCm', y='PetalLengthCm')

df_iris.info()
```

## Splitting

```python
sns.FacetGrid(df_iris, hue='Species', height=5).map(plt.scatter, 'PetalLengthCm', 'PetalWidthCm').add_legend()
```

```python
sns.FacetGrid(df_iris, hue='Species', height=5).map(sns.kdeplot, 'PetalLengthCm', 'PetalWidthCm').add_legend()
```

```python
df_iris.hist(edgecolor='black', linewidth=1.2)
fig = plt.gcf()
fig.set_size_inches(12, 9)
plt.show()
```

```python
df_iris.drop('Id', axis=1, inplace=True)
df_iris.head()
```

```python
plt.figure(figsize=(7, 4))
sns.heatmap(df_iris[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].corr(),annot=True,cmap='cubehelix_r')
plt.show()
```
